```python
# importing libraries
import numpy as np
import pandas as pd
from nltk import word_tokenize
```

```python
# lets read a csv file into a dataframe.
df = pd.read_csv(r'C:\Users\Chaitra.b.c\Downloads\bbc_text_cls.csv')
df.head()
```

```python
# Mapping word to index
word2index = {}

# tokenized_sentences
tokenized_sentences = []
idx = 0

for doc in df['text']:
    doc_as_idxs = []
    
    tokens = word_tokenize(doc.lower())
    
    for token in tokens:
        #print(token)
        if not word2index.get(token):
            word2index[token] = idx
            idx += 1
            
            doc_as_idxs.append(word2index[token])
    tokenized_sentences.append(doc_as_idxs)
```

```python
# the size of our vocabulary is 
V = len(word2index)
V
```

```python
# number of sentences
N = len(df['text'])
N
```

```python
# I need to create a tf matrix.
# Populate with term frequency ie. count of a term in a document.
# first lets create an empty matrix
tf_mat = np.zeros((N,V))
tf_mat
```

```python
tf_mat.shape
```

```python
# Now we need to update the count in the corresponding elements.
for i,doc_as_int in enumerate(tokenized_sentences):
    print(i,doc_as_int)
    for j in doc_as_int:
        if j < V:
            print(i,j)
            tf_mat[i,j] += 1
```

```python
# the last element in the matrix is at 2224x34761 as the starting element is at (0,0)
tf_mat.shape
```

```python
# calculating Doc_freq, It contains the number of documents each word appears in.
doc_freq = np.sum(tf_mat>0, axis = 0)
doc_freq
```

```python
# calculating the idf
IDF = np.log(N/doc_freq)
IDF.shape
```

```python
# printing the shape f the matrices
print(tf_mat.shape)
print(IDF.shape)
```

```python
#compute TF IDF
tf_idf = tf_mat * IDF
tf_idf
```

```python
index2word = {v:k for k,v in word2index.items()}
index2word
```

```python
i = np.random.choice(N)
row = df.iloc[i]

print('label : ', row['labels'])
print('text : ', row['text'].split('\n',1)[0])
print('Top 5 terms')

scores = tf_idf[i]
indices = (-scores).argsort()

for j in indices[:5]:
    print(index2word[j])
```

```python
from scipy.sparse import csr_matrix
from scipy.sparse import csr_array
```

```python
# term frequency CSR array generation from Scikit documentation.
docs = [["hello", "world", "hello"], ["goodbye", "cruel", "world"]]
indptr = [0]
indices = []
data = []
vocabulary = {}
for d in docs:
    for term in d:
        index = vocabulary.setdefault(term, len(vocabulary))
        indices.append(index)
        data.append(1)
    indptr.append(len(indices))

csr_array((data, indices, indptr), dtype=int).toarray()
```
