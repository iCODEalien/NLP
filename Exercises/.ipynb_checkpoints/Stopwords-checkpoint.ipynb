```python
import nltk
```

```python
# Downloading stop words
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
stopwords.words('german')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# A list of sentences
List_of_sentences = [
    'I hate them eggs',
    'I dont like Spirulina',
    'The albatross is a great bird',
    'Queen Elizabeth is GigaChad',
    'The Public Restrooms are some of the worst places on the planet'
]
```

```python
# creating a vectorizer object
vectorizer = CountVectorizer(analyzer = 'word', 
                             ngram_range = (1,1), 
                             stop_words = stopwords.words('english')) # here we can see the stopwords of english

# Generating the matrix
List_Matrix = vectorizer.fit_transform(List_of_sentences)

# converting the list martix and printing the array
print(List_Matrix.toarray())

# Feature names
print(vectorizer.get_feature_names_out())

# Stopwords used.
print(vectorizer.get_stop_words())
```
