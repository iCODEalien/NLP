```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk
import json
import matplotlib.pyplot as plt
```

```python
# Reading the csv file contains movies data
movies_df = pd.read_csv(r'C:\Users\Chaitra.b.c\Downloads/tmdb_5000_movies.csv')
movies_df.head()
```

```python
# getting the df info
movies_df.info()
```

```python
# missing value information
movies_df.isnull().sum()
```

```python
# For starters lets consider only the Genres and Keywords as the inputs for our tfidfVEctorizer.
# We need to process this json string.

#print('JSON to List : \n',json.loads(movies_df['keywords'][0]))

#convert them into one string of words.
keywords = [''.join(jo['name'].split(' ')) for jo in json.loads(movies_df['keywords'][0])]
print(' '.join(keywords))

genres = [''.join(jo['name'].split(' ')) for jo in json.loads(movies_df['genres'][0])]
print(' '.join(genres))
```

```python
# combining these steps into one function that returns 'keywords + genres'

def keywords_and_genres(df):
    keywords = ' '.join([''.join(jo['name'].split(' ')) for jo in json.loads(df['keywords'])])
    #print(' '.join(keywords))

    genres = ' '.join([''.join(jo['name'].split(' ')) for jo in json.loads(df['genres'])])
    #print(' '.join(genres))
    
    return '%s %s'%(keywords,genres)
```

```python
movies_df['Movie_detes_string'] = movies_df.apply(keywords_and_genres, axis = 1)
movies_df
```

```python
# creating a tfidfvectorizer object and fiting it with the movie details string
tfidf_vec =  TfidfVectorizer(max_features = 5000)
vec_matrix = tfidf_vec.fit_transform(movies_df['Movie_detes_string'])
vec_matrix
```

```python
# now we need a title to index series so we can get the corresponding index for each title
# This will also help us in finding the movie the user likes to get recommnedation for.

title2Index = pd.Series(movies_df.index, index = movies_df['original_title'])
title2Index
```

```python
# We get the index if the query.
query = 'Scream 3'
idx = title2Index[query]
```

```python
# Get the corresponding Vector.
query_vec = vec_matrix[idx]
query_vec
```

```python
# Cosine Similarity scores
scores = cosine_similarity(query_vec,vec_matrix)
scores.shape
```

```python
# We need to flatten these scores so it would become 1-D array
scores = scores.flatten()
scores.shape
```

```python
# the plot
plt.plot(scores)
```

```python
plt.plot(scores[(-scores).argsort()])
```

```python
recommended_index = (-scores).argsort()[1:6]
recommended_index
```

```python
movies_df['title'].iloc[recommended_index]
```

```python
def recommend(query):
    # We get the index if the query.
    idx = title2Index[query]
    
    # Get the corresponding Vector.
    query_vec = vec_matrix[idx]
    
    # find the Cosine Similarity
    scores = cosine_similarity(query_vec, vec_matrix)
    scores = scores.flatten()
    
    # Get the recommended movies indices
    recommend_indices = (-scores).argsort()[1:6]
    
    #returning the recommended movies
    return movies_df['original_title'].iloc[recommend_indices]
```

```python
recommend('Scream 3')
```

```python
recommend('Avatar')
```
