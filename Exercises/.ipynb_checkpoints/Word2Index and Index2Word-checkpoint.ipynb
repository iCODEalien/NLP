# Word to Index


```python
# Dictionary objects to map all the words in the text column in different ways.

# Appearance approach
word2index_appearance = {}
current_value = 1


# we have to get each value from text column.
for doc in df['text']:
    #print(len(doc))
    
    # use word tokenizer from nltk.
    # appearnance order
    tokens_appr = word_tokenize(doc)
    
    # Appearance approach.
    for token in tokens_appr:
        if not word2index.get(token):
            word2index[token] = current_value
            current_value += 1

word2index
    
```

```python
# Alphabetical order approach
word2index_alphaB_order = {}

all_words = []

for doc in df['text']:
    # extract tokens from document
    doc_words_list = word_tokenize(doc)
    
    # append them to all_words list
    all_words += doc_words_list

sorted_list = sorted(list(set(all_words)))
sorted_list_lwr = [element.lower() for element in sorted_list if element.isalpha()]

for idx,token in enumerate(sorted_list_lwr):
    word2index_alphaB_order[token] = idx

word2index_alphaB_order
```

```python
# Frequency order appraoch

from collections import Counter

all_words = []

for doc in df['text']:
    tokenized_sentence = word_tokenize(doc)
    all_words += tokenized_sentence

all_words = [element.lower() for element in all_words if element.isalpha()]
word_count = Counter(all_words)
word_count
```

```python
word_count_tuple = [(k,v) for k,v in dict(word_count).items()]
word_count_tuple
```

```python
sorted_word_count = sorted(word_count_tuple, key = lambda x: x[1])
sorted_word_count
```
